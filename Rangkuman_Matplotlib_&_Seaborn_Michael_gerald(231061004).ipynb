# Matplotlib
Matplotlib adalah sebuah pustaka (library) di Python yang digunakan untuk membuat visualisasi data dalam bentuk grafik 2D. Pustaka ini sangat populer di kalangan data scientist, insinyur, dan pengembang karena kemampuannya untuk membuat grafik yang menarik dengan mudah

beberapa fitur penting dari Matplotlib:

1.Beragam Tipe Grafik: Matplotlib dapat digunakan untuk membuat berbagai jenis grafik seperti garis (line plots), batang (bar plots), pie chart, scatter plot, histogram, dan banyak lagi

2.Highly Customizable: Anda dapat memodifikasi hampir semua elemen dari grafik yang dihasilkan, mulai dari warna, gaya garis, label, hingga ukuran font

3.Terintegrasi dengan NumPy dan Pandas: Matplotlib sering digunakan bersama NumPy dan Pandas untuk memvisualisasikan data dari array atau DataFrame

4.Modul pyplot: Sub-modul ini menyediakan interface yang mudah digunakan dan mirip dengan Matlab, sehingga membuat plotting lebih cepat dan sederhana


# Contoh Grafik

```python
#Grafik garis
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='sin(x)', color='pink')
plt.title('Grafik Sinus')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
#Garfik batang
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
plt.bar(categories, values, color='pink')
plt.title('Grafik Batang')
plt.xlabel('Kategori')
plt.ylabel('Nilai')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='red')
plt.title('Grafik Sebar')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

# Seaborn
Seaborn adalah pustaka visualisasi data yang ada di phyton dan berbasis Matplotlib yang lebih canggih dan elegan dimana Pustaka ini dirancang untuk memudahkan visualisasi statistik kompleks dengan gaya yang menarik dan tampilan yang siap untuk publikasi. Beberapa keunggulan utama Seaborn adalah:

1.Gaya Bawaan yang Lebih Menarik: Grafik yang dihasilkan oleh Seaborn sudah memiliki gaya yang estetis dan mudah dibaca, tanpa perlu banyak modifikasi.

2.Interaksi yang Kuat dengan Pandas: Seaborn bekerja sangat baik dengan data dalam format Pandas DataFrame, yang memudahkan manipulasi dan visualisasi data.

3.Grafik Statistik: Seaborn mempermudah pembuatan grafik statistik seperti distribusi data, hubungan antar variabel, heatmap, dan lain-lain.

4.Fitur Pemetaan Data Kategori: Seaborn memiliki kemampuan untuk membuat grafik berdasarkan data kategoris, seperti boxplot, barplot, dan violinplot.


# Contoh grafik

```python
#Grafik heatmap
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randint(1, 10, size=(3, 3))
sns.heatmap(data, annot=True, cmap='YlGnBu')
plt.title('Heatmap Data Matriks 3x3')

plt.show()
```

```python
#Grafik box plot
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data = [np.random.normal(size=100), np.random.normal(size=100, loc=2)]
sns.boxplot(data=data)
plt.title('Box Plot dari Dua Set Data')
plt.xlabel('Set Data')
plt.ylabel('Nilai')
plt.show()
```

```python
#Garfik scatter plot
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.random.rand(50)
y = 2 * x + np.random.rand(50)

data = pd.DataFrame({'X': x, 'Y': y})
sns.lmplot(x='X', y='Y', data=data, ci=None)
plt.title('Scatter Plot dengan Garis Regresi')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
